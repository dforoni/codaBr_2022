## <font color='red'> Dataframe </font>

**DataFrame** é uma estrutura de dados tabular bidimensional mutável em tamanho, e pode armazenar diferentes tipos de dados (int, float, string, date e etc.), com eixos rotulados (linhas e colunas), estrutura parecida com uma planilha de Excel/ Google Sheet.

dataframe.PNG


```python
# criando uma lista de textos, listas são indicadas por colchetes
lista = ['Coda.Br', 2022, 'Excel', 'Python']
type(lista)
```

**Nota:**

Uma lista é uma sequência de valores, assim como uma *string*.

Em uma *string*, os valores são caracteres, por outro lado, em uma lista, os valores podem ser de qualquer tipo. 

Os valores em lista são chamados de elementos ou às vezes chamamos de itens.

```python
# criando um dataframe com a lista criada anteriormente
dataframe = pd.DataFrame(lista, columns=['descrição'])
dataframe
```

## <font color='red'> Series </font>

**Seires** é uma matriz rotulada unidimensional capaz de conter qualquer tipo de dados (inteiros, strings (texto), números decimal, objetos Python, etc.), os rótulos dos eixos são chamados de índice. 

Cada coluna no *dataframe* é uma series, é igual uma coluna no Excel/ Google Sheet.

O índice começa no zero, lembrando que no Excel o índice começa no 1.

serie.PNG




```python
# criando uma series 
lista = ['Coda.Br', 2022, 'Excel', 'Python']
series = pd.Series(lista)
series
```

# <font color='red'> Queremos responder algumas perguntas, a partir dos dados de remuneração de funcionários do município de SP que carregamos. </font>

## 1. Qual foi o valor pago no mês de agosto de 2022?

```python
# utilizando a função sum
dados.sum(numeric_only=True)
```

```python
# formatando os valores float
pd.options.display.float_format = '{:,.2f}'.format
```

```python
# utilizando a função sum
dados.sum(numeric_only=True).sort_values(ascending=False)
```

## 2. Qual a quantidade de pessoas em cada status: Ativos, Aposentados e Pensionistas?

```python
# utilizando a função valeu counts 
dados['GRUPO'].value_counts()
```

```python
# utilizando a função valeu counts c/ proporção dos valores 
dados['GRUPO'].value_counts(normalize=True)
```

## 3. Qual o valor pago para a área da Educação?

### Selecionando por coluna

filtro.PNG

```python
# filtrando por coluna
dados.loc[:, 'ORGÃO']
```

### Selecionando por linhas

filtroRows.PNG

```python
# filtrando por coluna
dados.loc[1080815, :]
```

```python
# selecionando uma linha (DataFrame) - 2 dimensões (utilizando .loc)
dados.loc[[1080815], :]
```

```python
# criando um filtro com apenas os dados da SECRETARIA DA EDUCACAO
dados['ORGÃO'] == 'SECRETARIA DA EDUCACAO'
```

```python
filtro = dados['ORGÃO'] == 'SECRETARIA DA EDUCACAO'
filtro
```

```python
# filtrando o dataframe 
dados.loc[filtro]
```

```python
# salvando o dataframe em uma nova variável
dados_educacao = dados.loc[filtro]
dados_educacao
```

```python
# respondendo a pergunta:
dados_educacao['REMUNERAÇÃO DO MÊS'].sum()
```

## 4. Qual o custo de pagamento pela Secretária da Educação por Cargo?

```python
# utilizando a função groupby + agg
pvt = dados_educacao.groupby(['ORGÃO', 'CARGO'])['REMUNERAÇÃO DO MÊS'].agg(['sum', 'count'])
pvt
```
